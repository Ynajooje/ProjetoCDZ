```python
# Create a dictionary with the current column names and their Portuguese translations
new_column_names = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'porte_empresa'
}

# Rename the columns using the dictionary
df.rename(columns=new_column_names, inplace=True)

# Display the updated column names to confirm the changes
display(df.columns)
```

```python
# Create a dictionary to map the English experience levels to Portuguese
experience_level_map = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

# Apply the mapping to the 'nivel_experiencia' column
df['nivel_experiencia'] = df['nivel_experiencia'].replace(experience_level_map)

# Display the updated value counts to confirm the changes
display(df['nivel_experiencia'].value_counts())
```

```python
# Create a dictionary to map the English employment types to Portuguese
employment_type_map = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Apply the mapping to the 'tipo_emprego' column using replace
df['tipo_emprego'] = df['tipo_emprego'].replace(employment_type_map)

# Display the updated value counts to confirm the changes
display(df['tipo_emprego'].value_counts())
```

```python
# Create a dictionary to map the remote ratio to Portuguese terms
remote_map = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

# Apply the mapping to the 'remoto' column using replace
df['remoto'] = df['remoto'].replace(remote_map)

# Display the updated value counts to confirm the changes
display(df['remoto'].value_counts())
```

```python
#Criação de um dataframe teste para teste
df_salarios=pd.DataFrame({
    'nome': ["Ana Julia", "João", "Ana Cristina", "Julio", 'Klara'],
    'salario': [4000, np.nan, 3000, np.nan, 1000000]

    })
# Calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios ['salario'].fillna(df_salarios['salario'].mean().round(2))
# Calcula a mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios ['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas= pd.DataFrame ({
   'dia': ["Segunda", "Terça", "Quarta", "Quinta", 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas ["preenchido_ffill"] = df_temperaturas ["Temperatura"].ffill()
df_temperaturas
```
